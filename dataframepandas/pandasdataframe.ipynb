```python
point_df=pd.DataFrame({
    '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2st point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
})
```

```python
point_df
```

```python
def euclidean(row):
    pt_A=row['1st point']
    pt_B=row['1st point']
    return ((pt_A[0]-pt_B[1])**2 +(pt_A[1]-pt_B[0])**2)**0.5
```

```python
point_df['distance']= point_df.apply(euclidean, axis=1)
```
